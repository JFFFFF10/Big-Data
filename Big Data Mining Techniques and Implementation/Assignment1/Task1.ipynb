```python
import pandas as pd

data = pd.read_csv('adult.data', header=None, names=['age','workclass','fnlwgt','education','education-num','marital-status','occupation',
                                      'relationship','race','sex','capital-gain','capital-loss','hours-per-week',
                                      'native-country','income'])
# #To check for null and duplicated values
print(data.isnull().sum())
print(data.duplicated().sum())

# #Drop for duplicate values]
# data.drop_duplicates(inplace=True)

#To create a copy of the dataset for each question
question1 = data.copy()
question2 = data.copy()
question3 = data.copy()
question4 = data.copy()
question5 = data.copy()
```

### (1) Perform z-score normalization of the values in the attribute capital-gain. Show the mean and variance of the resulted values

```python
import pandas as pd

# Calculate mean and standard deviation
mean = question1["capital-gain"].mean()
std = question1["capital-gain"].std()

# Perform z-score normalization
question1["capital-gain"] = (question1["capital-gain"] - mean) / std

# Display mean and variance of the new values
print("Mean:", question1["capital-gain"].mean())
print("Variance:", question1["capital-gain"].var())
```

### (2) Choose one attribute with continuous values. Create five bins for this attribute such that the bins contain (approximately) equivalent numbers of records.

```python
#Give a name for each bins
labels = ['age_group1','age_group2','age_group3','age_group4','age_group5']

#Create an attribute to store the bins
question2['Age_catergory'] = pd.qcut(question2['age'], q=5, labels=labels)

question2
```

### (3) Apply one-hot-encoding to the attribute workclass.

```python
dummies = pd.get_dummies(question3['workclass'])
question3 = pd.concat([question3, dummies], axis=1)
question3.rename(columns={' ?':'unknown'}, inplace=True)
question3
```

### (4) Define at least one new attribute based on existing attribute, and explain your reason of defining it

To define a new attribute in relation to the income(class), I choose to add 'capital-gain' and 'capital-loss' to form 'total-capital' attribute. The reason behind creating this new attribute is that 'capital-gain' and 'capital-loss' are both financial attributes that provide information about a person's income. By creating a new attribute that sums these two factors, we can have a better understanding of a person's total capital and how it may relate to other factors in the dataset. This new attribute could be useful in the context of a financial analysis or prediction, where total capital is an important factor in determining a person's financial stability, creditworthiness or ability to save money.

```python
question4['total-capital'] = data['capital-gain'] + data['capital-loss']
question4
```

### (5) Find out and rank the correlations between the attribute hours-per-week and other continuous attributes.

```python
question5 = question5[['hours-per-week','age','fnlwgt','education-num','capital-gain','capital-loss']]

question5 = question5.corr(method='pearson').sort_values(by='hours-per-week',ascending=False)

question5['rank'] = question5['hours-per-week'].rank(method='first', ascending=False).astype('int')

question5
```
